```python
import gensim, logging
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
sentences = [['baseball','lg'], [ 'baseball','nc'], ['ss','baseball']]
# vector dimention : 2
def myhashfxn(obj):
    return hash(obj)%(2**32)

# size : 차원의 크기 (model.wv.syn0처럼 vector 시킬때 반환되는 vector의 크기)
# window : 학습시킬 두 단어 사이의 maximam distance
# min_count : 단어의 빈도수가 이 값보다 작은것은 출력x
# hashfxn : weight(가중치)를 랜덤하게 초기화 시켜주는 함수이다. 사용자 지정 OR Default가 주어진다.
model = gensim.models.Word2Vec(sentences, min_count=1,size=2, window=2, hashfxn=myhashfxn)
```

```python
for word in model.wv.index2word:
    print(word, model[word])
```

```python
# plt.plot(kind='scatter') 을 통해 산점도를 만들것이다.
# plt.text(x좌표, y좌표, 내용)
# iterrows() 는 DataFrame에서 index와 그 index에 따른 모든 column의 값을 필요로 할때 사용한다.(반복문에서 사용)
def drawPlot(v, x, y):
    v.plot(kind='scatter', x=x,y=y, figsize=(5,5))
    for idx, pos in v.iterrows():
        plt.text(pos[x], pos[y], idx )

        
# plt.gca(projection='3d') 는 3D 그래프를 그리는 함수.
# gca는 틀을 제공할 뿐이고 여기에 data는 scatter()로 추가해준다.
def draw3DPlot(v, x, y, z):
    fig = plt.figure(figsize=(8,8))
    threedee = fig.gca(projection='3d')
    threedee.scatter(v[x], v[y],v[z] )
    for idx, pos in v.iterrows():
        threedee.text(pos[x], pos[y],pos[z],  idx )
```

```python
# syn0 : 학습시킨 sentence의 단어들의 vector 값을 np.array로 반환 
# vocab, index2word : 학습 단어들의 index들을 반환

v = pd.DataFrame(model.wv.syn0, index=model.wv.index2word)
drawPlot(v, 0, 1)
```

```python
sentences = [['lg', 'baseball'], ['nc', 'baseball'], ['ss','baseball'], 
             ['lg','player','lee','good'],[ 'lg','player','ckm'],[ 'lg','player','jeehwan'], 
             ['ss','player','nbr'],[ 'ss','player','ajm'],[ 'ss','player','chw']             ]
# vector dimention : 3
model = gensim.models.Word2Vec(sentences, min_count=1,size=3, window=2)
```

```python
v = pd.DataFrame(model.wv.syn0, index=model.wv.vocab)
draw3DPlot(v, 0, 1, 2)
#drawPlot(v, 0, 1)
```

```python
# 저장 시킨 모델(Word2Vec) 를 불러들일때는 keyedVectors.load를 사용해야한다.
from gensim.models.keyedvectors import KeyedVectors
gn_model = KeyedVectors.load('m2v_rs.ml20.model')
```

```python
# most_simmilar() 를 통해 가장 postive와 가장 가깝고 negative와 가장 먼 index를 반환해준다.
gn_model.most_similar(positive=['Pretty Baby'], negative=['Mission: Impossible III'], topn=1)
```
